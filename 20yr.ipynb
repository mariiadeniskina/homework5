```python
import numpy as np
import random
n = 20
m = np.zeros((n, n + 1))
for i in range(n):
    m[i, i] = np.random.randint(100)
    m[i, n] = np.random.randint(100)
for i in range(n - 1):
    m[i + 1, i] = np.random.randint(100)
    m[i, i+1] = np.random.randint(100)
```

```python
def f(m):
    n = len(m)
    x = np.zeros(n)
    for i in range(n - 1):
        m[i+1, 0:] = m[i+1, 0:] - m[i,0:]/m[i,i]*m[i+1,i]
    for i in range(n-1, 0, -1):
        m[i-1,0:] = m[i-1,0:] - m[i,0:]*m[i-1,i]/m[i,i]
    for x in range(n):
        print(m[x,n]/m[x,x])
```
